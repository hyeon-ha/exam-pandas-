```python
from bs4 import BeautifulSoup
import requests
import re
```

```python
url = 'https://en.wikipedia.org/wiki/List_of_American_exchange-traded_funds'
resp = requests.get(url)
soup = BeautifulSoup(resp.text, 'lxml')
#print(soup)
rows = soup.select('div > ul > li')
#print(rows)
print(len(rows))
print(rows[20])
```

```python
etf_name = re.findall('^(.*) \(NYSE', rows[20].text)
print(etf_name)
etf_market = re.findall('\((.*)\|', rows[20].text)
print(etf_market)
etf_ticker = re.findall('NYSE Arca\|(.*)\)', rows[20].text)
print(etf_ticker)
```

```python
etfs = {}
etf_names = []
for row in rows:
    etf_name = re.findall('^(.*) \(NYSE', row.text)
    etf_market = re.findall('\((.*)\|', row.text)
    etf_ticker = re.findall('NYSE Arca\|(.*)\)', row.text)
    if (len(etf_name) > 0) & (len(etf_market) > 0) & (len(etf_ticker) > 0):
        etfs[etf_ticker[0]] = [etf_market[0], etf_name[0]]
print(etfs)
print(len(etfs))
```

```python
df_li = pd.DataFrame(etfs, index=['market', 'name'])
print(df_li.T)
```
