```python
import pandas as pd
from matplotlib import pyplot as plt
import matplotlib.pyplot as plt
import sklearn.datasets as datasets
from sklearn import datasets
```

```python
ds = datasets.load_iris()
print(type(ds))
print(ds.data.shape)
print(ds.target.shape)
print(ds)
```

```python
df = pd.DataFrame(ds['data'], columns=ds['feature_names'])
print(df.head())
```

```python
df['species2'] = [ds['target_names'][c] for c in ds['target']]
print(df[0:5])
```

```python
code_species_map = dict(zip(range(3), ds['target_names']))
df['species'] = [code_species_map[c] for c in ds['target']]
print(df[99:105])
```

```python
[code_species_map[c] for c in ds['target']]
code_species_list = []
for c in ds['target']:
    code_species_list.append(code_species_map[c])
print(code_species_list)
```

```python
sums_by_species = df.groupby('species').sum()
sums_by_species['sepal length (cm)'].plot(
    kind='pie', fontsize=10, autopct='%0.1f',
    explode=(0.1, 0.1, 0.1), shadow=True,
    startangle=90, legend=True)
plt.ylabel('Sepal Length', labelpad=40, horizontalalignment='right')
plt.title('Iris classified as the length of a sepal',
          fontsize=15)

plt.show()
```

```python
sums_by_species
```

```python
fig = sums_by_species.plot(kind='pie', subplots=True,
                    layout=(2,2), figsize=(10,10),
                    legend=True)
plt.suptitle('Total Measurements according to species, by Species')
plt.show()
```

```python
sums_by_species['sepal width (cm)'].plot(
    kind='bar', fontsize=10)
plt.ylabel('Sepal Width', horizontalalignment='right')
plt.title('Iris classified as the width of a sepal',
         fontsize=10)
plt.show()
```

```python
sums_by_species.plot(kind='bar', subplots=True,
    layout=(4,1), figsize=(3, 12), legend=False)
plt.suptitle('Total Measurements according to species, by Species')
plt.show()
```

```python
for spec in df['species'].unique():
    forspec = df[df['species']==spec]
    forspec['petal length (cm)'].plot(
        kind='hist',alpha=0.8,label=spec)
plt.legend(loc='upper right')
plt.suptitle('Histogram by petal length')
plt.show()
```

```python
petal_length = df['petal length (cm)']
print(type(petal_length))
print(petal_length.head())
average = petal_length.mean()
std = petal_length.std()
median = petal_length.quantile(0.5)
percentile25 = petal_length.quantile(0.25)
percentile75 = petal_length.quantile(0.75)
clean_avg = petal_length[(petal_length>percentile25)
                        & (petal_length < percentile75)].mean()
print('average :', average)
print('std :', std)
print('median :', median)
print('percentile25 :', percentile25)
print('percentile75 :', percentile75)
print('clean average :', clean_avg)
```

```python
df.plot(kind='scatter', x='sepal length (cm)',
       y='sepal width (cm)')
plt.title('Length vs Width')
plt.show()
```

```python
colors = ['r', 'g', 'b']
markers = ['.', '*', '^']
fig, ax = plt.subplots(1,1)
for i, spec in enumerate(df['species'].unique()):
    ddf = df[df['species']==spec]
    ddf.plot(kind='scatter', 
             x='sepal length (cm)',
             y='sepal width (cm)',
             alpha=0.5, s=10*(i+1), ax=ax,
             color=colors[i], marker=markers[i],
             label=spec)
plt.legend()
plt.show()
```
